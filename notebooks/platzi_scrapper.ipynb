```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
service = webdriver.chrome.service.Service(executable_path= 'C:/chromedriver_win32/chromedriver.exe')
browser = webdriver.Chrome(options=options, service= service)
```

```python
url = 'https://www.platzi.com'

browser.get(url)
html = browser.page_source

print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')

courses_links = soup.select('.SchoolsList-school')
courses = [course['href'] for course in courses_links]

for course in courses:
    print(course)
```
